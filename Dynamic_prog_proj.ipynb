### dp - двумерная динамика по количеству выбранных типов. Значения - максимальная стоимость для данной точки (кол-во типов, макс вес)

```python
dp = np.array([[0]*(M+1)]*(N+1))
dp
```

```python
for i in range(1, N+1):    #brute force of types 
    for w in range(1, M+1):     #w - available weight for i types
        dp[i][w] = dp[i-1][w]
        for x_i in range(0, w//goods[0][i] + 1):    #x_i-number of goods
            dp[i][w] = max(dp[i][w], dp[i-1][w-x_i*goods[0][i]] + goods[1][i]*x_i) #Bellman equation
            print(i, ' ', w, ' ', x_i,dp[i-1][w], ' ', dp[i-1][w-x_i*goods[0][i]] + goods[1][i]*x_i, dp[i][w])
        print()
```

### восстановление ответа

```python
answer = np.array([0]*N)
```

```python
def restore_answer(i, w):
    if  goods[1][i] != 0 and dp[N][M]/goods[1][i] == dp[N][M]//goods[1][i] and goods[1][i] != 0:
        answer[i-1] = dp[N][M]//goods[1][i]
        return 0
    elif dp[i][w] == 0:
        return 0
    elif dp[i][w] == dp[i-1][w]:
        print('*')
        restore_answer(i-1, w)
    else:
        x_i = (dp[i][w] - dp[i-1][w])//goods[1][i]
        restore_answer(i-1, w - goods[0][i]*x_i)
        answer[i-1] = x_i 
```
